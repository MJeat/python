## 4. Verify Library Installation

Let's check that all the libraries were installed correctly and can be imported.

```python
print("üîç Verifying library installations:\n")

libraries_to_check = {
    'numpy': 'Numerical Python',
    'pandas': 'Data Analysis Library',
    'matplotlib': 'Plotting Library',
    'sklearn': 'Scikit-learn (Machine Learning)',
    'seaborn': 'Statistical Visualization',
}

successful_imports = 0
total_imports = len(libraries_to_check)

for library, description in libraries_to_check.items():
    try:
        if library == 'sklearn':
            import sklearn
            version = sklearn.__version__
        else:
            imported_lib = __import__(library)
            version = getattr(imported_lib, '__version__', 'Unknown')
        
        print(f"‚úÖ {library:12} ({description}) - Version: {version}")
        successful_imports += 1
    except ImportError:
        print(f"‚ùå {library:12} ({description}) - Failed to import")

print(f"\nüìä Import Summary: {successful_imports}/{total_imports} libraries imported successfully")

if successful_imports == total_imports:
    print("üéâ All libraries imported successfully! You're ready for the course.")
else:
    print("‚ö†Ô∏è  Some libraries failed to import. Please check the error messages above.")
```
